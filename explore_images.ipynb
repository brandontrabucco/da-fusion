```python
from PIL import Image
import numpy as np
from glob import glob
```

```python
def image_grid(imgs, rows, cols):
    assert len(imgs) == rows*cols

    w, h = imgs[0].size
    grid = Image.new('RGB', size=(cols*w, rows*h))
    grid_w, grid_h = grid.size
    
    for i, img in enumerate(imgs):
        grid.paste(img, box=(i%cols*w, i//cols*h))
    return grid
```

```python
def show_n(n, path,seed=0):
    ims = sorted(glob(path+'*.png'))
    rng = np.random.default_rng(seed)
    ids = rng.permutation(len(ims))[:n]
    open_ims = [Image.open(ims[i]) for i in ids]
    return image_grid(open_ims[:], 1, n)
```

```python
path_pooled = './results/250_qtr/synthetic_images/spurge_five_fold_pooled/'
path_imspec = './results/250_qtr/synthetic_images/spurge_five_fold_imspec/'
```
